```python
import math
import numpy as np
```

```python
def gradiente_f(x,y):
    grad_x = 6*x+3*y-1
    grad_y = 3*x+4*y-2
    return (grad_x,grad_y)

def ponto_com_gradiente(x0,y0,alpha):
    grad_x, grad_y = gradiente_f(x0,y0)
    x1 = x0-alpha*grad_x
    y1 = y0-alpha*grad_y
    return (x1,y1)

alpha = .1
x0 = 0
y0 = 0

n = True
pontos = []

while n:
    gradiente_x, gradiente_y = gradiente_f(x0,y0)
    x0, y0 = ponto_com_gradiente(x0,y0,alpha)
    pontos.append((x0,y0))
    if np.sqrt(gradiente_x**2+gradiente_y**2) < 10**-5:
        n = False

print(pontos[-1])
```
