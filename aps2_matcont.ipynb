```python
import math

def grad_f(x, y):
    df_dx = 6*x + 3*y - 1
    df_dy = 3*x + 4*y - 2
    return df_dx, df_dy

def descida_gradiente_f(alpha, start_x, start_y, precision=1e-5):
    x, y = start_x, start_y
    iterations = 0
    
    while True:
        grad_x, grad_y = grad_f(x, y)
        
        next_x = x - alpha * grad_x
        next_y = y - alpha * grad_y
        
       
        distance = math.sqrt((next_x - x)**2 + (next_y - y)**2)
        
        iterations += 1
        x, y = next_x, next_y
        
      
        if distance < precision:
            break
            
        
        if iterations > 20000:
            return float('inf'), float('inf'), iterations
            
    return x, y, iterations

 
ponto_inicial = (0, 0)
prec = 1e-5 


alphas = [0.1, 0.15, 0.2, 0.3, 0.5]

print("--- Tarefa 1: Resultados para f(x,y) ---")
print(f"Ponto Inicial: {ponto_inicial}, Precisão: {prec}\n")

for alpha in alphas:
    min_x, min_y, iters = descida_gradiente_f(alpha, ponto_inicial[0], ponto_inicial[1], prec)
    
    if math.isinf(min_x):
        print(f"Passo (Alpha) = {alpha}:")
        print(f"  Resultado: DIVERGIU (passo muito grande).")
        print(f"  Iterações: > {iters}\n")
    else:
        print(f"Passo (Alpha) = {alpha}:")
        print(f"  Resultado (Ponto Mínimo): ({min_x:.6f}, {min_y:.6f})")
        print(f"  Iterações: {iters}\n")
```

```python
import math

def grad_g(x, y):

    raiz = math.sqrt(x**2 + y**2 + 1)
    exp = math.exp(-(y**2))
    
    dg_dx = (x / raiz) + (2 * x * exp) + (2 * x - 4)
    dg_dy = (y / raiz) - (2 * y * (x**2) * exp)
    
    return dg_dx, dg_dy


def gradient_descent_g(alpha, start_x, start_y, precision=1e-5):
    x, y = start_x, start_y
    iterations = 0
    
    while True:
        grad_x, grad_y = grad_g(x, y)
        
        next_x = x - alpha * grad_x
        next_y = y - alpha * grad_y
        
        distance = math.sqrt((next_x - x)**2 + (next_y - y)**2)
        
        iterations += 1
        x, y = next_x, next_y
        
        if distance < precision:
            break
            
        if iterations > 10000:
            return float('inf'), float('inf'), iterations
            
    return x, y, iterations


print("--- Tarefa 2: Resultados para g(x,y) ---")
alpha_g = 0.1 
prec = 1e-5


ponto_inicial_1 = (0, 0)
min_x_1, min_y_1, iters_1 = gradient_descent_g(alpha_g, ponto_inicial_1[0], ponto_inicial_1[1], prec)
print(f"Iniciando em {ponto_inicial_1}: Ponto mínimo 1 ({min_x_1:.6f}, {min_y_1:.6f}) em {iters_1} iterações.")


ponto_inicial_2 = (-1, 0)
min_x_2, min_y_2, iters_2 = gradient_descent_g(alpha_g, ponto_inicial_2[0], ponto_inicial_2[1], prec)
print(f"Iniciando em {ponto_inicial_2}: Ponto mínimo 2 ({min_x_2:.6f}, {min_y_2:.6f}) em {iters_2} iterações.")
```

```python
import math


def grad_h(x, y):
    exp1 = math.exp(-(x**2) - (y**2))
    common_exp_val = -(x**2) - (y**2) + 4*x + 6*y - 13
    exp2 = math.exp(common_exp_val)
    
    dh_dx = -8 * x * exp1 + (12 - 6*x) * exp2 - (2*x / 7)
    dh_dy = -8 * y * exp1 + (18 - 6*y) * exp2 - (y / 6)
    
    return dh_dx, dh_dy


def gradient_ascent_h(alpha, start_x, start_y, precision=1e-5):

    x, y = start_x, start_y
    iterations = 0
    
    while True:
        grad_x, grad_y = grad_h(x, y)
        
        
        next_x = x + alpha * grad_x
        next_y = y + alpha * grad_y
        
        distance = math.sqrt((next_x - x)**2 + (next_y - y)**2)
        
        iterations += 1
        x, y = next_x, next_y
        
        if distance < precision:
            break
            
        if iterations > 10000:
            return float('inf'), float('inf'), iterations
            
    return x, y, iterations



print("--- Tarefa 3: Resultados para h(x,y) ---")
alpha_h = 0.1
prec = 1e-5

ponto_inicial_1 = (0, 0)
max_x_1, max_y_1, iters_1 = gradient_ascent_h(alpha_h, ponto_inicial_1[0], ponto_inicial_1[1], prec)
print(f"Iniciando em {ponto_inicial_1}: Ponto máximo 1 ({max_x_1:.6f}, {max_y_1:.6f}) em {iters_1} iterações.")


ponto_inicial_2 = (2, 3)
max_x_2, max_y_2, iters_2 = gradient_ascent_h(alpha_h, ponto_inicial_2[0], ponto_inicial_2[1], prec)
print(f"Iniciando em {ponto_inicial_2}: Ponto máximo 2 ({max_x_2:.6f}, {max_y_2:.6f}) em {iters_2} iterações.")
```

```python
import math

def f(x, y):
    return 3*x**2 + 3*x*y + 2*y**2 - x - 2*y



def grad_f(x, y):
  
    df_dx = 6*x + 3*y - 1
    df_dy = 3*x + 4*y - 2
    return df_dx, df_dy


def gradient_descent_variable_step(start_x, start_y, precision=1e-5):

    x, y = start_x, start_y
    iterations = 0
    

    beta = 0.5

    while True:
        grad_x, grad_y = grad_f(x, y)
        
        
        grad_norm = math.sqrt(grad_x**2 + grad_y**2)
        
        
        if grad_norm < precision:
            break
            
        
        alpha = 1.0 

        current_f_val = f(x, y)
        
        while True:
            next_x = x - alpha * grad_x
            next_y = y - alpha * grad_y
            
         
            if f(next_x, next_y) < current_f_val:
                break 
            
           
            alpha = alpha * beta
            
         
            if alpha < 1e-10:
                break

        x, y = next_x, next_y
        iterations += 1
        
        if iterations > 1000:
            return float('inf'), float('inf'), iterations

    return x, y, iterations


print("--- Tarefa Desafio: Passo Variável para f(x,y) ---")

start_point = (0, 0) 
prec = 1e-5          

min_x_var, min_y_var, iters_var = gradient_descent_variable_step(start_point[0], start_point[1], prec)

print(f"Iniciando em {start_point}, Precisão: {prec}")
print(f"Passo Variável: Ponto mínimo ({min_x_var:.6f}, {min_y_var:.6f}) "
      f"em {iters_var} iterações.")
```
